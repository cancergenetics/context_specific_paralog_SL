```python
cell_lines = pd.DataFrame(target_pairs['DepMap_ID'].unique()).rename(columns={0:"DepMap_ID"})
```

```python
# save pathnames of csv PPI files
folder = '/Users/narod/Documents/GitHub/context_specific_SL_prediction/1_feature_calculation/annotations/input_files/ranked_essentiality'

ranked_files = list()
ranked_filenames = list()

for filename in os.listdir(folder):
    if filename.endswith('.csv'):
    
        base_filename = os.path.split(filename)[1]
        ranked_filenames.append(base_filename.split(".")[0])
        
        file_directory = os.path.join(folder, filename)
        ranked_files.append(file_directory)

    elif filename.endswith('.parquet'):
        base_filename = os.path.split(filename)[1]
        ranked_filenames.append(base_filename.split(".")[0])

        file_directory = os.path.join(folder, filename)
        ranked_files.append(file_directory)
    else:
        continue
```

```python
def process_ranked_ess_files(ranked_files, cell_lines):

    results = []

    for file in ranked_files:
        ranked_df = pd.read_csv(file, index_col=0)
        filtered_ranked_df = ranked_df[ranked_df.index.isin(cell_lines['DepMap_ID'])]
        filtered_ranked_df = filtered_ranked_df.reset_index(drop=False).rename(columns={"index":"DepMap_ID"})
        
        # Convert wide format to long format
        melt_ranked_df = pd.melt(filtered_ranked_df, id_vars=["DepMap_ID"], var_name="entrez_id", value_name="value")
        melt_ranked_df['entrez_id'] = melt_ranked_df['entrez_id'].astype(int)
        melt_ranked_df = melt_ranked_df.rename(columns={'value': os.path.basename(file).split(".")[0]})
        results.append(melt_ranked_df)
        print(f"Processed {file}")

    return results
```

```python
list_of_ranked_files = process_ranked_ess_files(ranked_files, cell_lines)
```

```python
from functools import reduce

ranked_merged_df = reduce(lambda left, right: pd.merge(left, right, on=['DepMap_ID', 'entrez_id'], how='left'), list_of_ranked_files)
```

```python
ranked_merged_df[:3]
```

```python
target_pairs_annotated = pd.merge(target_pairs, ranked_merged_df.rename(columns={'entrez_id':'A1_entrez', 'ranked_essentiality':'A1_rank', 'ranked_zessentiality':'zA1_rank'}), 
                     on=['DepMap_ID', 'A1_entrez'], how='left')
target_pairs_annotated = pd.merge(target_pairs_annotated, ranked_merged_df.rename(columns={'entrez_id':'A2_entrez', 'ranked_essentiality':'A2_rank', 'ranked_zessentiality':'zA2_rank'}), 
                     on=['DepMap_ID', 'A2_entrez'], how='left')
```

```python
target_pairs_annotated['max_ranked_A1A2'] = target_pairs_annotated[['A1_rank', 'A2_rank']].max(axis=1)
target_pairs_annotated['min_ranked_A1A2'] = target_pairs_annotated[['A1_rank', 'A2_rank']].min(axis=1)

target_pairs_annotated['z_max_ranked_A1A2'] = target_pairs_annotated[['zA1_rank', 'zA2_rank']].max(axis=1)
target_pairs_annotated['z_min_ranked_A1A2'] = target_pairs_annotated[['zA1_rank', 'zA2_rank']].min(axis=1)
```
